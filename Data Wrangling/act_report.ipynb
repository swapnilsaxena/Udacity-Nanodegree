### Analysis and Insights of WeRateDogs Twitter archive

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import the Twitter archive master CSV file into a DataFrame

time_df = pd.read_csv("C:\\Users\\Swapnil Saxena\\Desktop\\Data Engineering\\2.0 Data Wrangling\\Twitter Project\\twitter_archive_master.csv")
```

```python
time_df['timestamp'] = pd.to_datetime(time_df['timestamp'])
```

```python
data = time_df.tweet_id.groupby([time_df["timestamp"].dt.year, time_df["timestamp"].dt.month]).count()
ax = data.plot(kind = 'line', title='WeRateDogs tweeting trend over time')
ax.set_xlabel("(Year, Month)")
ax.set_ylabel("Number of Tweets")
ax.set_ylim(0, )
plt.savefig('tweet_trend')
```

WeRateDogs, a Twitter account started in 2015, tweeted a lot in its starting days. For example, it posted
around 300 tweets in the month of November 2015, which increased to around 350 tweets in exactly the next
month. However, its tweeting activity reduced drastically in the subsequent months, averaging between 50 to
100 since April, 201

```python
ax = time_df.rating_numerator.value_counts().sort_index().plot(kind='bar', title = 'Dog Rating distribution')
ax.set_xlabel("Rating out of 10")
ax.set_ylabel("Number of Dogs")
ax.set_yticks([0, 50, 100, 150, 200, 250, 300, 350, 400, 450])
plt.savefig('rating_dist')
```

Most of the dogs are rated 12/10 (i.e. 455 out of 1994 dogs). Excluding the 2 rating outliers (420 and 1776),
the highest rating received by any dog is 14/10. However, only 2% (i.e. 37) dogs got this rating. So, we can
say that WeRateDogs is probably very conservative when it comes to rating a dog the highest.

```python
# Create rating_ration variable by dividing the rating numerator by the deonominator to normalize scores which are not 
# out of 10
time_df['rating_ratio'] = time_df['rating_numerator']/time_df['rating_denominator']
```

```python
time_df['retweet_count'].plot(color = 'red', label='Retweets')
time_df['favorite_count'].plot(color = 'blue', label='Favorites')
plt.style.use('seaborn-darkgrid')
plt.legend(loc='upper left')
plt.xlabel('Tweet timestamp')
plt.ylabel('Count')
plt.title('Retweets and favorites over time')
plt.figure(figsize = (10,50))
plt.savefig('retweets_favorites.png')
plt.show()
```

The above plot shows an increasing trend of retweeting and favoriting tweets posted by WeRateDogs' since
November, 2015. However, the favoriting trend shows a much higher increase with time as compared to the
retweeting trend.

```python
time_df['source'].value_counts().plot(kind='barh', figsize=(11,5), title='Most used Twitter source').set_xlabel("Number of Tweets")
plt.savefig('twitter_source')
```

#### WeRateDogs has posted 98% of the tweets from iPhone
Out of the 1987 tweets, 1949 were posted from iPhone. Just a confirmation that the only mobile device
WeRateDogs uses to post tweets is an iPhone.
